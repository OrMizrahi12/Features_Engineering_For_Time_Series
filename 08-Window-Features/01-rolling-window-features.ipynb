# Computing rolling windows features using Pandas

```python
# Create a copy of the data.
df = data.copy()
```

We want to compute the rolling:
- mean
- standard deviation
- median absolute deviation

The median absolute deviation is not implemented natively in Pandas so we have to create it as a custom metric ourselves.

Let's create a function to compute the median absolute deviation which is defined as: $MAD = median(|x_i - median(x)|)$

```python
# Create a custom metric for the MAD
def mad(x):
    return np.median(np.abs(x - np.median(x)))
```

We'll use a window size of 24 hours to smooth over daily seasonality. Let's compute the rolling window features!

```python
result = (
    df["demand"]               # the column we want to rolling & shift.
    .rolling(window=24)        # Pick window size.
    .agg(["mean", "std", mad]) # Pick statistics.
    .shift(freq="1H")          # Lag by 1 hour to avoid data leakage.
)  

result = result.add_prefix("demand_window_24_")

result
```

Let's join this back to the original dataframe.

```python
# Add the rolling features to the original data frame
df = df.join(result, how="left")
df
```

The `min_periods` argument can be used to allow smaller window sizes to be applied at the start of the time series. This is one method to reduce the amount of missing data at the start of the time series. Alternatively, we could drop the missing data altogether.

```python
result = (
    df["demand"]
    .rolling(window=24, min_periods=1) # Pick window size.
    .agg(["mean", "std", mad]) # Pick statistics.
    .shift(freq="1H") # Lag by 1 hour to avoid data leakage.
)  

result = result.add_prefix("demand_window_24_min_periods_")

df = df.join(result, how="left")
# show those columns
df[["demand", "demand_window_24_mean", "demand_window_24_min_periods_mean"]]
```

# Using Feature-engine

Let's see how we can extract window features using the Feature-engine library.

```python
from feature_engine.timeseries.forecasting import WindowFeatures
```

```python
# Create a copy of the data.
df = data.copy()
```

```python
# Create window features using the transformer.
transformer = WindowFeatures(
    variables=["demand", "temperature"],
    functions=["mean", "std"],
    window=[24, 24 * 7, 24 * 365], # Day, week, year.
    freq="1H",
)

df = transformer.fit_transform(df)
df
```

Let's look at the rolling window mean of the demand.

```python
ax = df.filter(
    regex="demand_.*?_mean", # `.*?` means any number of any characters.
    axis=1  # Filter by column names.
).plot(figsize=[14, 7])

ax.set_title("Rolling window mean of electricity demand")
ax.set_ylabel("Electricity demand")
ax.set_xlabel("Time")
```

We can see how different window sizes have smoothed over different seasonalities. This allows the model to make use of behaviours seen at different time scales (e.g., if the daily average of the demand is greater than the average of the past year then perhaps this can influence future values of our target variable).

# Using sktime

Let's see how we can extract window features using the sktime library.

```python
from sktime.transformations.series.summarize import WindowSummarizer
```

```python
df = data.copy()
```

```python
transformer = WindowSummarizer(
    lag_feature={
        "lag": [1, 2, 3],  # Lag features.
        "mean": [[1, 24], [1, 24*7]],  # [[lag, window size]]
        "std": [[1, 24], [3, 24*7]],
        "mad": [[1, 12]],  
       # mad: [[1, 12]],  # Can pass custom functions.[make an err]
    },
    target_cols=["demand"],
)

result = transformer.fit_transform(df)
result
```

```python
df = result.join(df["demand"], how="left")
df
```

So now we've shown how to create rolling window features using different libraries.
